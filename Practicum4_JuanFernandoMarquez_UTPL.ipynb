```python
##Filtrado de datos 1
##Por logica deducimos que genero (gender) y edad (age) teniendo datos desconocidos NA (NaN) no aportan mucho a nuestro 
##analisis de datos por lo que procedemos a actualizar nuestro dataset sin estos valores.
##Gender = NA
##Age = NA
```

```python
df_New=df[(df["gender"].notnull())&(df["age"].notnull())]
```

```python
##Filtrado de datos 2
##Los valores de gender: male y female no son manipulables, por lo que procedemos a convertirlos en valores
##que se puedan manejar 
## 0 = femenino
## 1 = masculino 
```

```python
df_New['gender'].replace(['female','male'],[0,1],inplace=True)
```
